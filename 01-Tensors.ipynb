```python
import numpy as np

# Define the x and y ranges
x = np.array([0, 1, 2])
y = np.array([0, 1, 2, 3])

# Create the meshgrid
xx, yy = np.meshgrid(x, y)

# Print the results
print("xx:\n", xx)
print("\nyy:\n", yy)
```

```python
grid = np.stack([xx, yy], axis=-1)
print(grid)
```

```python
grid.shape
```

```python
points = [[0.4, 0.3], [0.6, 0.7]]
print(points)
points = np.array(points)
print(points)
points.shape
```

```python
grid = grid[..., None, :]  # Shape: (res, res, 1, 2)
print(grid)
grid.shape
```

```python
points = points[None, None, :, :]  # Shape: (1, 1, n_points, 2)
print(points)
```

```python
points.shape
```

```python
distances = np.sqrt(np.min(np.sum((grid - points)**2, axis=-1), axis=-1))
print(distances)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.imshow(distances)
```
